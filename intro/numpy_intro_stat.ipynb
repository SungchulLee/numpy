```python
import numpy as np
a = np.array(2, dtype=np.float64)
b = np.array([1,2,3])
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[[1,2,3],[4,5,6]]])
print(a, a.shape, a.dtype)
print(b, b.shape, b.dtype)
print(c, c.shape, c.dtype)
print(d, d.shape, d.dtype)
```

```python
# shape 바꾸기

import numpy as np

a = np.array([[[1,2,3],[4,5,6]]]).reshape((3,2)) # (1, 2, 3) ---> (3, 2)
print(a.shape)
print(a)
```

```python
# dtype 바꾸기

import numpy as np

#a = np.array([[[1,2,3],[4,5,6]]]).astype(np.float32) # int64 ---> float32
a = np.array([[[1,2,3],[4,5,6]]], dtype=np.float32)
print(a.dtype)
print(a)
```

# indexing and slicing

```python
a = np.array([[1,2,3], [4,5,6]]) # (2,3)
print(a) # (2,3)
# print(a[0,1]) # index ---> dim reduce : ()
# print(a[0,:]) # (3,)
# print(a[0]) # a[0,:] (3,)
# print(a[:,1]) # (2,)
print(a[0,1]) # ()
print(a[0:1,1]) # (1,)
print(a[0,1:2]) # (1,)
print(a[0:1,1:2]) # (1,1)
```

```python
import numpy as np
a = np.random.normal(size=(10000,))

x = np.linspace(-3,3,100)
y = np.exp(-x**2/2) / np.sqrt(2*np.pi)

import matplotlib.pyplot as plt
plt.plot(x, y, alpha=0.5, color='r')
plt.hist(a, bins=np.linspace(-3,3,100), density=True)
plt.show()
```
