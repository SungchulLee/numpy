```python
n = 10
x = np.linspace(-10, 10, n) # (100,)
y = np.log(np.abs(x)) # (100,) 
s = np.linspace(1, 10, n) * 100

import matplotlib.pyplot as plt
#plt.plot(x, y, '-*r')
plt.plot(x, y,
         linestyle='-.',
         linewidth=5,
         color='#EC1052',
         marker='P',
         markersize=20,
         markeredgecolor='b',
         markeredgewidth=3,
         markerfacecolor='c')
#plt.plot(x, y, 'o', markersize=20)
#plt.scatter(x, y, s=s)
```

```python
n = 100
x = np.random.normal(size=(n,)) # (100,)
epsilon = np.random.normal(size=(n,))
noise_level = 2
noise = noise_level * epsilon
y_obs = 1 + 2 * x + noise
y_true = 1 + 2 * x
plt.plot(x, y_obs, 'ob')
plt.plot(x, y_true, 'r')
```

$$
f(x)=\frac{1}{\sqrt{2\pi}} e^{-x^2/2}
$$

```python
n = 10000
normal_sample = np.random.normal(size=(n,)) # (10000,)
x = np.linspace(-3,3,100)
y = np.exp(-x**2/2) / np.sqrt(2*np.pi)
plt.hist(normal_sample, bins=30, density=True)
plt.plot(x, y, 'r--', alpha=0.5)
```
