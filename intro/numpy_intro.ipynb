```python
n = 100
x = np.random.normal(size=(n,)) # (100,)
epsilon = np.random.normal(size=(n,))
noise_level = 2
noise = noise_level * epsilon
y_obs = 1 + 2 * x + noise
y_true = 1 + 2 * x
plt.plot(x, y_obs, 'ob')
plt.plot(x, y_true, 'r')
```

# componentwise operations vs matrix operations

```python
a = np.eye(2)
b = np.array([[1,2],[3,4]], dtype='float')
# c1 = a * b # componentwise operations
# c2 = np.multiply(a, b)
# c3 = a.__mul__(b)
c1 = a @ b # matrix operations
c2 = np.matmul(a, b)
c3 = a.__matmul__(b)
print(c1)
print(c2)
print(c3)
```

# broadcasting

```python
a = np.array([1,2,3]) # (3,)
b = 2 # (,) --- > (3,) 분신술 사용 가능
#b = np.array([2]) # (1,) --- > (3,) 분신술 사용 가능
#b = np.array([2,2]) # (2,) --- > (3,) 분신술 사용 불가능
c = a + b
print(a)
print(b)
print(c)
```

```python
# # () ---> 분신술 사용 가능
# # (1,) ---> 분신술 사용 가능
# # (2,) ---> 분신술 사용 불가능

# # ok
# 3,2,1,1,2
#         2
    
# # ok
# 3,2,1,1,2
#   2,1,1,2
    
# # ok
# 3,2,1,5,2
#       1,2
    
# # not ok
# 3,2,1,1,2
#   5,1,1,2
    
# # ok
# 3,1,1,1,2
#   5,1,1,2   
```
