```python
import numpy as np
x = np.array([[1,2,3]])
y = np.array([1,1,1])
z = x + y # componentwise addition
print(x, type(x), x.shape, x.dtype)
print(y, type(y), y.shape, y.dtype)
print(z, type(z), z.shape, z.dtype)
```

```python
import numpy as np
x = np.array([[1],[2],[3]]) # (3, 1)
y = np.array([[1,2,1],[1,2,1],[1,2,1]]) # (3, 3)
z = x + y # componentwise addition
print(x, type(x), x.shape, x.dtype)
print(y, type(y), y.shape, y.dtype)
print(z, type(z), z.shape, z.dtype)
```

```python
import numpy as np
x = np.array([[1],[2],[3]]) # (3, 1)
y = np.array([[1,2,1],[1,2,1],[1,2,1]]) # (3, 3)
z = x - y # componentwise op
print(x, type(x), x.shape, x.dtype)
print(y, type(y), y.shape, y.dtype)
print(z, type(z), z.shape, z.dtype)
```

```python
import numpy as np
x = np.array([[1],[2],[3]]) # (3, 1)
y = np.array([[1,2,1],[1,2,1],[1,2,1]]) # (3, 3)
z = x * y # componentwise op
print(x, type(x), x.shape, x.dtype)
print(y, type(y), y.shape, y.dtype)
print(z, type(z), z.shape, z.dtype)
```

```python
import numpy as np
x = np.array([[1],[2],[3]]) # (3, 1)
y = np.array([[1,2,1],[1,2,1],[1,2,1]]) # (3, 3)
z = x / y # componentwise op
print(x, type(x), x.shape, x.dtype)
print(y, type(y), y.shape, y.dtype)
print(z, type(z), z.shape, z.dtype)
```

```python
import numpy as np
x = np.array([[1,2,1],[1,2,1],[1,2,1]]) # (3, 3)
y = np.array([[1],[2],[3]]) # (3, 1)
z = x @ y # matrix multiplication
print(x, type(x), x.shape, x.dtype)
print(y, type(y), y.shape, y.dtype)
print(z, type(z), z.shape, z.dtype)
```

```python
import numpy as np
x = np.array([[1,2,1],[1,2,1],[1,2,1]]) # (3, 3)
y = np.sin(x) # componentwise op
y = np.cos(x) # componentwise op
y = np.tan(x) # componentwise op
y = np.exp(x) # componentwise op
y = np.log(x) # componentwise op
print(x, type(x), x.shape, x.dtype)
print(y, type(y), y.shape, y.dtype)
```
