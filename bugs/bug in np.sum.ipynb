```python
import time 
tic = time.time()
n = 10000
s_n = int( n * (n + 1) * ( 2*n + 1 ) / 6 )
print(s_n)
toc = time.time()
print(toc-tic)
```

```python
import time 
tic = time.time()
n = 10000
s_n = 0
for i in range(1, n+1):
    s_n += i**4
print(s_n)
toc = time.time()
print(toc-tic)
```

```python
import time 
tic = time.time()
n = 10000
lst = []
for i in range(1, n+1):
    lst.append(i**4)
s_n = sum(lst)
print(s_n)
toc = time.time()
print(toc-tic)
```

```python
import numpy as np
import time 
tic = time.time()
n = 10000
lst = np.power(np.arange(1, n+1), 4)
# s_n = lst.sum()
s_n = np.sum(lst)
print(s_n)
toc = time.time()
print(toc-tic)
```
