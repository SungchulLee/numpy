Flip $p$-coin $n$ times and report the result:
```
n = 5   # number of coin clips 
p = 0.7 # prob of head
```

````python
import numpy as np

n = 5   # number of coin clips 
p = 0.5 # prob of head

for _ in range(n):
    if np.random.uniform() > 1 - p: 
        print('H', end='\t')
    else:
        print('T', end='\t')
````

````python
import numpy as np
import matplotlib.pyplot as plt

n = 5   # number of coin clips 
p = 0.5 # prob of head

img_head = plt.imread('head.png')
img_tail = plt.imread('tail.png')

fig, axes = plt.subplots(nrows=1, ncols=n, figsize=(100, 50))
for i in range(n):
    if np.random.uniform() > 1 - p:
        axes[i].imshow(img_head)
    else:
        axes[i].imshow(img_tail)
    axes[i].axis('off')
plt.show()
````
